# Exercício de Histograma

## Enunciado

O ojetivo desta tarefa é aolicar a distribuição de frequência utilizando o atributo age da base da dados do censo

*   Carregue a base de dados *census.csv*
*   Faça teste utilizando o parâmentro **bins** para visualizar a distribuição dos dados




## Resolução

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/Estatística para Ciência de Dados e Machine Learning/Bases de dados/census.csv')
```

```python
dataset.head()
```

```python
def config(title, xlabel, ylabel):
  plt.title(title)
  plt.xlabel(xlabel)
  plt.ylabel(ylabel)
  plt.show()
```

```python
dataset['age'].plot.hist(bins=5)
config('Distribuição de Frequência', 'Idade', 'Frequência')
```

```python
dataset['age'] = pd.cut(dataset['age'], bins=[0,17,25,40,60,90], labels=['Jovem', 'Adolescente', 'Jovem Adulto', 'Adulto', 'Idoso'])
```

```python
dataset.head()
```

```python
dataset['age'].unique()
```

```python
dataset['age'].value_counts().plot(kind='bar')
config('Distribuição de Frequência por Categoria', 'Categoria de Idade', 'Frequência')
```
