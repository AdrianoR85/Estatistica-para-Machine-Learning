## Porcentagem

### Enunciado

O objetivo desta tarefa é encontrar algum tipo de relação entre o atributo ***education*** e o atributo ***income*** da base de dados do censo.

- Carregue o arquivo census.csv

- Crie um novo dataframe com o pandas utilizando somente esses dois campos

- Você pode utilizar o comando groupby do pandas para agrupar os dados, ou seja, a quantidade de registros baseado nesses dois atributos. Acesse aqui a documentação

- Identifique o nível de educação que possuem as maiores e menores rendas. Lembre-se que nesta base de dados a renda assume somente os valores >50 e <=50

### Resolução

```python
import pandas as pd
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/Estatística para Ciência de Dados e Machine Learning/Bases de dados/census.csv')[['income','education']]
```

```python
dataset
```

```python
dataset2 = dataset.groupby(['education','income']).size().reset_index(name='counts')
```

```python
dataset2
```

```python
education_total = dataset2.groupby('education')['counts'].transform('sum')
dataset2['%_education'] = (dataset2['counts'] / education_total) * 100

grand_total = dataset2['counts'].sum()

dataset2['%_overall'] = (dataset2['counts'] / grand_total) * 100
```

## Coeficientes e Taxas

### Enunciado

O objetivo deste exercício é calcular o coeficiente e a taxa de evasão utilizando a tabela abaixo. Faça os cálculos manuais e depois implemente em Python

image.png

### Resolução

```python
import pandas as pd
```

```python
data = {'ano': ['1', '2','3','4', 'total'],
        'matriculas_marco': [70,50,47,23,190],
        'matriculas_novembro': [65,48,40,22,175]}

df = pd.DataFrame(data)
df
```

```python
df['coeficiente'] = (df['matriculas_marco'] - df['matriculas_novembro']) / df['matriculas_marco']
df
```

```python
df['taxa_evasao'] = df['coeficiente'] * 100
df
```

### Insight



* **Tendência Geral de Redução (Exceto Ano 3):** Excluindo o pico do ano 3, há uma tendência geral de melhora na retenção, com as taxas caindo do ano 1 (7.14%) para o ano 2 (4.00%) e para o ano 4 (4.34%).

* **Picos Significativos na Evasão:** O ponto mais imediato a se observar é o salto na taxa de evasão do ano 2 para o ano 3. A taxa aumentou de 4% para quase 15% (14.893617%).
